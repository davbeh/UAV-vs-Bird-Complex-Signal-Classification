```python
import numpy as np
import matplotlib.pyplot as plt

# RBF kernel k(x, x0) = exp(-gamma * (x - x0)^2)
def rbf(x, x0, gamma):
    return np.exp(-gamma * (x - x0)**2)

# Axis & reference point
x = np.linspace(-5, 5, 1000)
x0 = 0.0

# Gamma values (the higher gamma, the narrower the peak)
gammas = [0.1, 0.5, 1.0, 2.0, 5.0]

plt.figure(figsize=(8, 4.5))
for g in gammas:
    plt.plot(x, rbf(x, x0, g), label=fr"$\gamma={g}$")

# Appearance
#plt.title("RBF kernel for different $\gamma$ (center $x_0=0$)")
plt.xlabel("x")
plt.ylabel(r"$k(x,y)$")
plt.ylim(-0.05, 1.05)
#plt.grid(True, alpha=0.3)
#plt.legend()

plt.tight_layout()
#plt.savefig("/RBF_Kernel.png", transparent=True)
plt.show()
```
